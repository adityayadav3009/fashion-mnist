```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.metrics import accuracy_score
```

```python
train = pd.read_csv("fashion-mnist_train.csv")
train.head()
```

```python
train.describe()
```

```python
for col in train.columns[1:] :
     train[col] =train[col]/255 
```

```python
layer_0 =tf.keras.layers.Dense(units=len(train.columns[1:]),input_shape=[784])
layer_1 =tf.keras.layers.Dense(units=128 ,activation=tf.nn.relu )
layer_2 =tf.keras.layers.Dense(units=10 ,activation=tf.nn.softmax )
```

```python
model =tf.keras.Sequential([layer_0,layer_1,layer_2] )
```

```python
model.compile(loss="sparse_categorical_crossentropy" ,optimizer="adam" ,metrics=["accuracy"] )
```

```python
model.fit(train[train.columns[1:]] ,train.label ,epochs=20)
```

```python
test =pd.read_csv("fashion-mnist_test.csv")
predicted =model.predict(test[train.columns[1:]])
```

```python
ans=[]
for i in range(len(predicted)) :     
     ans.append(np.argmax(predicted[i]))
     
test["predicted_label"] = ans
```

```python
accuracy_score(test["label"] ,test["predicted_label"])
```
